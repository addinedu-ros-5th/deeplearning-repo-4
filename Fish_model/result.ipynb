```python
import os
import pandas as pd

folder_path = '/Users/yohan/Desktop/dev_ws/git_ws/deeplearning-repo-4/Fish_model/Fish_segment/result_csv'
file_names = ['/Users/yohan/Desktop/dev_ws/git_ws/deeplearning-repo-4/Fish_model/Fish_segment/result_csv/results1.csv',
              '/Users/yohan/Desktop/dev_ws/git_ws/deeplearning-repo-4/Fish_model/Fish_segment/result_csv/results2.csv',
              '/Users/yohan/Desktop/dev_ws/git_ws/deeplearning-repo-4/Fish_model/Fish_segment/result_csv/results3.csv']  # 가져오고 싶은 파일 이름 리스트
output_file_name = 'combine.csv'  # 새로 생성할 파일 이름

batch_sizes = [32, 64, 128]  # 새로 추가된 부분

dfs = []
for i, file_name in enumerate(file_names):
    file_path = os.path.join(folder_path, file_name)
    df = pd.read_csv(file_path)
    # 'epoch'를 행으로, 'train/box_loss'를 열로 변경
    df = df.set_index('epoch').T
    # 'batch' 칼럼 추가
    df['batch'] = batch_sizes[i]
    # 'batch' 칼럼을 맨 왼쪽으로 이동
    df = df[['batch'] + [col for col in df.columns if col != 'batch']]
    dfs.append(df)

combined_df = pd.concat(dfs, ignore_index=False)

# 새로운 CSV 파일 생성
output_file_path = os.path.join(folder_path, output_file_name)
combined_df.to_csv(output_file_path, index=True)

combined_df
```
